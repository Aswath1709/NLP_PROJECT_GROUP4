```python
# Funtion to change review helpfulness ratio to float
def convert_ratio_to_float(ratio_str):
    if pd.isna(ratio_str) or '/' not in ratio_str:
        return None
    numerator, denominator = ratio_str.split('/')
    try:
        return float(numerator) / float(denominator)
    except ZeroDivisionError:
        return None
```

```python
# Apply convert_ratio_to_float function to review/helpfulness column
merged_df['review/helpfulness'] = merged_df['review/helpfulness'].apply(convert_ratio_to_float)
```

```python
# Define a function to clean text in our data
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def clean_text(text):
    # change text to lowercase
    text = text.lower()
    # filter out non-alphabetical characters
    text = re.sub(r'[^a-z]', ' ', text)
    # split text
    words = text.split()
    # Stem text and filter out words that are not in stopwords set
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
# Apply cleaning to all text data in our set
merged_df['description'] = merged_df['description'].apply(clean_text)
merged_df['review/summary'] = merged_df['review/summary'].apply(clean_text)
merged_df['review/text'] = merged_df['review/text'].apply(clean_text)
```

```python
# Create a function to change review/time to datetime data type
def convert_published_date(date_str):
    if pd.isna(date_str):
        return pd.NaT
    try:
        return pd.to_datetime(date_str)
    except:
        return pd.NaT

merged_df['publishedDate'] = merged_df['publishedDate'].apply(convert_published_date)
```

We are creating a sample dataset from the original dataset to tag sentiments to reviews using BERT classifier, since running the model on origina dataset wold take much time. So, we tag sentiments to sample dataset, run setiment classification model (implemented after tagging sample dataset) on this dataset and use the classification model to predict the labels iin the original dataset

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader
from torch.nn.functional import softmax
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)  # Assuming binary sentiment (positive/negative)

# Data Preprocessing
max_length = 128  # You can adjust this based on your requirements

def preprocess_text(text):
    inputs = tokenizer(text, max_length=max_length, return_tensors='pt', truncation=True, padding='max_length')
    return inputs

# Apply sentiment analysis to each row in the DataFrame
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

def analyze_sentiment(text):
    inputs = preprocess_text(text)
    inputs = {key: value.to(device) for key, value in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)

    probs = softmax(outputs.logits, dim=1).detach().cpu().numpy()[0]
    sentiment_label = 'positive' if probs[1] < probs[0] else 'negative'

    return sentiment_label

# Apply sentiment analysis to each row in the DataFrame
sampled_df['sentiment'] = sampled_df['review/text'].apply(analyze_sentiment)

# Display the updated DataFrame
print(sampled_df[['review/text', 'sentiment']])
```
